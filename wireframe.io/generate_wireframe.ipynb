```python
<VSCode.Cell id="1" language="markdown"># Generate Caribbean Beach Wireframe PNG
This notebook creates a PNG wireframe for a luxury Caribbean tropical beach company and saves it inside the separate `wireframe.io` folder.</VSCode.Cell>
<VSCode.Cell id="2" language="python">import os
from PIL import Image, ImageDraw, ImageFont

print('Imports ready')</VSCode.Cell>
<VSCode.Cell id="3" language="python">output_dir = os.path.join(os.getcwd(), 'wireframe.io')
os.makedirs(output_dir, exist_ok=True)
print('Output folder:', output_dir)</VSCode.Cell>
<VSCode.Cell id="4" language="python">WIDTH, HEIGHT = 1400, 900
BACKGROUND = (245, 248, 250)
ACCENT = (34, 94, 86)
TEXT = (30, 41, 51)
MUTED = (107, 114, 128)
LINE = (209, 213, 219)

image = Image.new('RGB', (WIDTH, HEIGHT), BACKGROUND)
draw = ImageDraw.Draw(image)

try:
    header_font = ImageFont.truetype('arialbd.ttf', 40)
    label_font = ImageFont.truetype('arial.ttf', 20)
except IOError:
    header_font = ImageFont.load_default()
    label_font = ImageFont.load_default()

print('Canvas defined')</VSCode.Cell>
<VSCode.Cell id="5" language="python">def draw_box(bounds, outline=LINE, width=2, fill=None):
    draw.rectangle(bounds, outline=outline, width=width, fill=fill)

# Header
header_bounds = (40, 40, WIDTH - 40, 120)
draw_box(header_bounds)
draw.text((60, 50), 'Caribbean Coastal Retreats', fill=ACCENT, font=header_font)

# Navigation
nav_labels = ['Home', 'Destinations', 'Experiences', 'About', 'Contact']
nav_x = WIDTH - 420
for label in nav_labels:
    draw.text((nav_x, 60), label, fill=MUTED, font=label_font)
    nav_x += 110

# Hero section
hero_bounds = (40, 150, WIDTH - 40, 510)
draw_box(hero_bounds)
hero_image_bounds = (60, 170, WIDTH - 420, 490)
draw_box(hero_image_bounds)
draw.text((90, 260), 'Hero Image', fill=MUTED, font=label_font)

hero_text_x = WIDTH - 360
draw.text((hero_text_x, 190), 'Experience the Luxury of Caribbean Shores', fill=TEXT, font=header_font)
draw.text((hero_text_x, 240), 'Curated private villas, oceanfront escapes, and bespoke beachside moments.', fill=MUTED, font=label_font)

draw_box((hero_text_x, 310, hero_text_x + 220, 360), outline=ACCENT)
draw.text((hero_text_x + 20, 320), 'Explore Villas', fill=ACCENT, font=label_font)

draw_box((hero_text_x + 230, 310, hero_text_x + 450, 360), outline=ACCENT)
draw.text((hero_text_x + 250, 320), 'Plan Escape', fill=ACCENT, font=label_font)

# Feature cards
card_y = 540
card_width = (WIDTH - 140) // 3
card_height = 220
for i, title in enumerate(['Private Beach Villas', 'Exclusive Yacht Tours', 'Wellness Retreats']):
    x0 = 40 + i * (card_width + 30)
    draw_box((x0, card_y, x0 + card_width, card_y + card_height))
    draw_box((x0 + 20, card_y + 20, x0 + card_width - 20, card_y + 110), outline=LINE)
    draw.text((x0 + 25, card_y + 40), 'Image', fill=MUTED, font=label_font)
    draw.text((x0 + 20, card_y + 130), title, fill=TEXT, font=label_font)
    draw.text((x0 + 20, card_y + 160), 'Premium travel experience', fill=MUTED, font=label_font)

# Footer
draw_box((40, 780, WIDTH - 40, 880))
footer_labels = ['About', 'Destinations', 'Services', 'Contact']
for i, heading in enumerate(footer_labels):
    x0 = 50 + i * 320
    draw.text((x0, 795), heading, fill=ACCENT, font=label_font)
    draw.text((x0, 825), 'Link 1', fill=MUTED, font=label_font)
    draw.text((x0, 855), 'Link 2', fill=MUTED, font=label_font)
    draw.text((x0, 885), 'Link 3', fill=MUTED, font=label_font)

print('Wireframe drawn')</VSCode.Cell>
<VSCode.Cell id="6" language="python">output_path = os.path.join(output_dir, 'wireframe.png')
image.save(output_path)
output_path</VSCode.Cell>
<VSCode.Cell id="7" language="python">from PIL import Image
from IPython.display import Image as DisplayImage

DisplayImage(filename=os.path.join(output_dir, 'wireframe.png'))</VSCode.Cell>
```
